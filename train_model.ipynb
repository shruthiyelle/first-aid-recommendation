```python
# @title Default title text
# Import libraries
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import pickle

# Create a dataset
data = {
    "symptom": [
        "burn", "cut", "choking", "sprain", "fever",
        "cold", "cough", "headache", "allergic reaction",
        "nosebleed", "poisoning", "fainting", "electric shock"
    ],
    "recommendation": [
        "Cool with water for 10 mins. Don't apply ice or creams.",
        "Clean wound, apply pressure, cover with sterile bandage.",
        "Perform Heimlich maneuver. Call emergency services if needed.",
        "Rest, ice, compress, elevate.",
        "Stay hydrated, rest, take paracetamol.",
        "Rest, drink warm fluids, use over-the-counter remedies.",
        "Stay hydrated, use honey or warm drinks.",
        "Rest in a dark room, take ibuprofen.",
        "Take antihistamine. If breathing issues, use epinephrine and call emergency services.",
        "Lean forward, pinch nose, hold for 10 minutes.",
        "Call emergency services immediately.",
        "Lay down, raise legs, loosen clothing.",
        "Ensure away from source, check breathing, call emergency."
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Train the model
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['symptom'])
y = df['recommendation']

model = MultinomialNB()
model.fit(X, y)

# Save model and vectorizer
pickle.dump(model, open("model.pkl", "wb"))
pickle.dump(vectorizer, open("vectorizer.pkl", "wb"))

print("✅ Model trained and saved!")
```
